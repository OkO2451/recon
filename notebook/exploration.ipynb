```python
import os
import subprocess

def invoquer_gobuster(url_target):
    # Ensure url_target is not empty
    if not url_target:
        print("Url/Domain (-u) must be specified.")
        return []

    # Construct the absolute path to the wordlist
    script_dir = os.path.dirname(os.path.abspath(__file__))
    wordlist_path = os.path.join(script_dir, '../data/subDomaine.txt')

    # Check if the wordlist file exists and is not empty
    

    try:
        # Construct the gobuster command
        command = [
            "gobuster", "dir",
            "-w", wordlist_path,
            "-u", url_target,
            "-s", "200,204,301,302,307,401,403",
            "-b", ""
        ]
        
        # Debug output: print the command
        print(f"Running command: {' '.join(command)}")

        # Execute the gobuster command and capture output
        result = subprocess.run(
            command,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )

        # Debug output: print result codes and output
        print(f"Command exited with code: {result.returncode}")
        print(f"stdout: {result.stdout}")
        print(f"stderr: {result.stderr}")

        # Handle errors
        if result.returncode != 0:
            print("Gobuster encountered an error:")
            print(result.stderr)
            return []

        # Split the output into a list of lines
        output = result.stdout.split('\n')

        # Remove empty lines
        output = [line for line in output if line.strip()]

        return output

    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Example usage
url_target = "http://example.com"
result = invoquer_gobuster(url_target)
print(result)
```

```python
import os
import subprocess
from IPython.display import display, HTML

def invoquer_gobuster(url_target):
    # Ensure url_target is not empty
    if not url_target:
        print("Url/Domain (-u) must be specified.")
        return []

    # Attempt to find the current notebook's directory
    try:
        from pathlib import Path
        from IPython.core.getipython import get_ipython

        # Get the current notebook's full path
        notebook_path = Path(get_ipython().starting_dir)
        wordlist_path = notebook_path / '../data/subDomaine.txt'
    except Exception as e:
        print(f"Could not determine notebook's directory: {e}")
        return []

    # Check if the wordlist file exists and is not empty
    if not os.path.isfile(wordlist_path):
        print(f"Wordlist file does not exist: {wordlist_path}")
        return []
    if os.path.getsize(wordlist_path) == 0:
        print(f"Wordlist file is empty: {wordlist_path}")
        return []

    try:
        # Construct the gobuster command
        command = [
            "gobuster", "dir",
            "-w", str(wordlist_path),
            "-u", url_target,
            "-s", "200,204,301,302,307,401,403",
            "-b", ""
        ]
        
        # Debug output: print the command
        print(f"Running command: {' '.join(command)}")

        # Execute the gobuster command and capture output
        result = subprocess.run(
            command,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )

        # Debug output: print result codes and output
        print(f"Command exited with code: {result.returncode}")
        print(f"stdout: {result.stdout}")
        print(f"stderr: {result.stderr}")

        # Handle errors
        if result.returncode != 0:
            print("Gobuster encountered an error:")
            print(result.stderr)
            return []

        # Split the output into a list of lines
        output = result.stdout.split('\n')

        # Remove empty lines
        output = [line for line in output if line.strip()]

        return output

    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Example usage
url_target = "http://example.com"
result = invoquer_gobuster(url_target)
print(result)
```
